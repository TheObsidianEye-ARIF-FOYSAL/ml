```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
import numpy as np
```

```python
x, y = make_regression(n_samples=4, n_features=1, n_informative=1, n_targets=1,  noise=80, random_state=13)
```

```python
plt.scatter(x, y)
```

```python
# Lets apply OLS
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(x, y)

LinearRegression(copy_X=True, fit_intercept=True, n_jobs=None)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(x, y)
plt.plot(x, reg.predict(x), color='red')
plt.show()
```

```python
# apply Gradient Decent assuming slop is constant m = 78.35
# asume the sarting value for intercept b = 0
y_pred = (78.35 * x + 0).reshape(4)
plt.scatter(x, y)
```

```python
plt.scatter(x, y)
plt.plot(x, reg.predict(x), color='red', label='OLS')
plt.plot(x, y_pred, color='green', label='Gradient Descent, b = 0')
plt.legend()
plt.show()
```

```python
m = 78.35
b = 0

loss_slope = -2*np.sum(y - m*x.ravel() - b)
loss_slope
```

```python
# Lets take learning rate = 0.1
lr = 0.1

step_size = loss_slope*lr
step_size
```

```python
# Calculating the new intercept
b_new = b - step_size
b_new
```

```python
y_pred1 = (78.35 * x + b_new).reshape(4)
plt.scatter(x, y)
plt.plot(x, reg.predict(x), color='red', label='OLS')
plt.plot(x, y_pred1, color='green', label='b = {:.2f}'.format(b_new))
plt.plot(x, y_pred, color='blue', label='b = 0')
plt.legend()
plt.show()
```

```python
# Iteration 2
loss_slope = -2*np.sum(y - m*x.ravel() - b_new)
loss_slope
step_size = loss_slope*lr
step_size
```

```python
b_new = b_new - step_size
b_new
```

```python
y_pred2 = (78.35 * x + b_new).reshape(4)
plt.scatter(x, y)
plt.plot(x, reg.predict(x), color='red', label='OLS')
plt.plot(x, y_pred1, color='green', label='b = {:.2f}'.format(b_new))
plt.plot(x, y_pred2, color='black', label='b = {:.2f}'.format(b_new))
plt.plot(x, y_pred, color='blue', label='b = 0')
plt.legend()
plt.show()
```

```python
# One code
b = -100
m = 78.35
lr = 0.01
epochs = 100
for i in range(epochs):
    loss_slope = -2*np.sum(y - m*x.ravel() - b)
    step_size = loss_slope*lr
    b = b - step_size
    y_pred = m*x + b
    plt.plot(x, y_pred, label='b = {:.2f}'.format(b))
    print(f"Iteration {i+1}: b = {b}")
```
